## Part a
Use the URL identified above, and write code that fetches eBay's search result page for "samsung tv". Save the result to a file. Name the file as "ebay_samsung_tv_01.htm".

```python
# open the url and store the content
url = "https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2380057.m570.l1312.R1.TR0.TRC0.A0.H0.TRS2&_nkw=samsung+tv&_sacat=0"
html = urllib.request.urlopen(url).read()

# create a function to save the html file with a assigned name
def saveHtml(filename, file_content):
    with open(filename + ".html", "wb") as f:
        f.write(file_content)
        
# save the the search result to 'ebay_samsung_tv_01.html'    
saveHtml("ebay_samsung_tv_01", html)
```

## Part b 
Take your code in (a) and write a loop that will download the first 10 pages of search results. Save each of these pages. Name the files as "ebay_samsung_tv_<page_number>.htm", where you replace "<page_number>" with a TWO digit page number corresponding to search page number (for examples, "ebay_samsung_tv_05.htm" for the 5th page of search). IMPORTANT: Each page request needs to be followed by at least a 10-second pause! Remember, you want your program to mimic your behavior as a human and help you make good purchasing decisions.

```python
# set the url pattern that determine the page number
page_pat = "&_pgn="

for i in range(1,11):
    # get the url for the ith page
    new_url = url + page_pat + str(i)
    
    # get the file name for ith page
    if(i < 10):
           page_num = "0" + str(i)
    else:  page_num = str(i)
    file_name = "ebay_samsung_tv_" + page_num
    
    # get the content of ith page
    html_content = urllib.request.urlopen(new_url).read()
    
    # save the ith html file
    saveHtml(file_name, html_content)
    
    
```

## Part c
Write a separate piece of code that loops through the pages you downloaded in (b) (i.e. the saved files) and opens and parses them into an internal object of your programming language of choice (For example, if you are using Python the recommended object is BeautifulSoup object or for Java, jsoup object is recommended). The object should allow you to select elements from it. Next, find the sponsored items on each search result page and print their URL to the screen.

```python
# Set the Sponsored pattern
pattern = ".*S.*P.*O.*N.*S.*O.*R.*E.*D"
# create the list to store the sponsored links for ten pages
full_sponsored = []

# set up a loop to iterate each page  
for i in range(1, 11):
    # find the file name for each page
    if(i < 10):
           page_num = "0" + str(i)
    else:  page_num = str(i)
    file_name = "ebay_samsung_tv_" + page_num + ".html"
    # store the page content in 'soup'
    f = open(file_name)
    soup = BeautifulSoup(f, 'html.parser')

    # create a list to store the urls of the sponsored items in this page
    sponsored = []
    # class .s-item__link store item information
    # get the infomation of all items
    items = soup.select(".s-item__link")
    
    # set up a loop to iterate each searching item
    for item in items:
        # class s-item__title--tagblock store the word "Sponsored"
        sp = item.select(".s-item__title--tagblock")[0]
        txt = sp.span.text
        result = bool(re.search(pattern, txt))
        # to determine if it is sponsored
        if (result == True):
            link = item.get('href')
            # append the link of sponsored item in this page
            sponsored.append(link)
    
    # append the links of sponsored items in each page to a list
    full_sponsored.extend(sponsored)          
         
            
            
```

```python
# print the links of the sponsored items
print(full_sponsored)
```
